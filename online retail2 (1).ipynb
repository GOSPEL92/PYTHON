```python
#Customer segmentation,helps classify customers based on their purchasing behavior, 
#allowing for targeted marketing and personalized engagement.
import datetime

# Ensure InvoiceDate is in datetime format
df_cleaned["InvoiceDate"] = pd.to_datetime(df_cleaned["InvoiceDate"])

# Define a reference date (assumed dataset's latest date + 1 day)
latest_date = df_cleaned["InvoiceDate"].max() + datetime.timedelta(days=1)

# Compute RFM metrics
rfm_df = df_cleaned.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (latest_date - x.max()).days,  # Recency
    "InvoiceNo": "nunique",  # Frequency (unique transactions)
    "TotalSpend": "sum"  # Monetary (total revenue)
})

# Rename columns
rfm_df.rename(columns={"InvoiceDate": "Recency", "InvoiceNo": "Frequency", "TotalSpend": "Monetary"}, inplace=True)

# Display results
print(rfm_df.head())
```

Key Insights on Customer segmentation
Helps classify high-value customers for targeted campaigns. 
Identifies churn riskcustomers who havent purchased in a while.
Allows personalized promotions based on customer behavior.

```python
#apply K-Means Clustering to segment customers into groups based on their purchasing behavior. 
#This helps identify high-value customers, potential churners, and casual buyers.
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Standardize RFM values
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_df)

# Find optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

#  Elbow plot to determine the best number of clusters
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal Clusters")
plt.show()

# Apply K-Means with chosen number of clusters (e.g., 3)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
rfm_df["Cluster"] = kmeans.fit_predict(rfm_scaled)

# Visualize customer clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=rfm_df["Recency"], y=rfm_df["Monetary"], hue=rfm_df["Cluster"], palette="coolwarm")
plt.xlabel("Recency (Days Since Last Purchase)")
plt.ylabel("Monetary Value ($)")
plt.title("Customer Segmentation using K-Means")
plt.show()
```

Key Insights for Customer Clustering
Helps identify loyal customers vs. potential churners.
Useful for personalized marketing and retention strategies.
 Clustering reveals spending behavior and purchase patterns
